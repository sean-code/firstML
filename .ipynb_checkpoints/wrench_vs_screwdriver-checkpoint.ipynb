```python
# --- Cell 1: Imports & basic paths ---
from pathlib import Path
import os, shutil
from typing import List, Tuple

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from tqdm import tqdm

# Torch / torchvision
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from torchvision.models import resnet18, ResNet18_Weights

# Metrics
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.model_selection import train_test_split

# Paths and constants
RAW_DIR    = Path("data/raw")
IMAGES_DIR = Path("data/images")
TRAIN_DIR  = Path("data/train")
VALID_DIR  = Path("data/valid")
TARGET_CLASSES = ["wrench", "screwdriver"]
VALID_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".gif", ".webp"}

for p in [RAW_DIR, IMAGES_DIR, TRAIN_DIR, VALID_DIR]:
    p.mkdir(parents=True, exist_ok=True)

print("Using device:", "cuda" if torch.cuda.is_available() else "cpu")
```

```python
# --- Cell 2: Data preparation (copy & split) ---
# This will:
# 1) Collect images for each class from data/raw/**/<class name> folders (case-insensitive).
# 2) Copy them to data/images/<class>/
# 3) Create an 80/20 split into data/train/<class>/ and data/valid/<class>/

def find_class_dirs(root: Path, class_name: str):
    matches = []
    for d in root.rglob("*"):
        if d.is_dir() and class_name.lower() in d.name.lower():
            matches.append(d)
    return matches

def collect_images(src_dirs):
    files = []
    for sd in src_dirs:
        for fp in sd.rglob("*"):
            if fp.is_file() and fp.suffix.lower() in VALID_EXTS:
                files.append(fp)
    return files

def safe_copy(src: Path, dst: Path):
    dst.parent.mkdir(parents=True, exist_ok=True)
    stem, suffix = src.stem, src.suffix.lower()
    final = dst
    i = 1
    while final.exists():
        final = dst.with_name(f"{stem}_{i}{suffix}")
        i += 1
    shutil.copy2(src, final)

# 1) Gather images into data/images/<class>/
for cls in TARGET_CLASSES:
    cls_dirs = find_class_dirs(RAW_DIR, cls)
    cls_out = IMAGES_DIR / cls
    cls_out.mkdir(parents=True, exist_ok=True)
    if cls_dirs:
        print(f"Found {len(cls_dirs)} folder(s) that look like '{cls}':")
        for d in cls_dirs:
            print("  -", d)
        imgs = collect_images(cls_dirs)
        print(f"Copying {len(imgs)} images -> {cls_out}")
        for img in tqdm(imgs, desc=f"Copy {cls}"):
            safe_copy(img, cls_out / img.name)
    else:
        print(f"WARNING: No folders found for '{cls}'. Make sure {RAW_DIR}/.../{cls}/ has images.")

# 2) Train/valid split (80/20), clearing old splits first
if TRAIN_DIR.exists(): shutil.rmtree(TRAIN_DIR)
if VALID_DIR.exists(): shutil.rmtree(VALID_DIR)
TRAIN_DIR.mkdir(parents=True, exist_ok=True)
VALID_DIR.mkdir(parents=True, exist_ok=True)

def train_valid_split(class_dir: Path, train_dir: Path, valid_dir: Path,
                      valid_frac: float = 0.2, seed: int = 42):
    files = [p for p in class_dir.glob("*") if p.suffix.lower() in VALID_EXTS]
    if not files:
        print(f"WARNING: No images found in {class_dir}")
        return 0, 0
    rng = np.random.RandomState(seed)
    idx = np.arange(len(files))
    rng.shuffle(idx)
    split = int(len(files) * (1 - valid_frac))
    train_idx, valid_idx = idx[:split], idx[split:]
    for i in train_idx:
        dst = train_dir / class_dir.name / files[i].name
        safe_copy(files[i], dst)
    for i in valid_idx:
        dst = valid_dir / class_dir.name / files[i].name
        safe_copy(files[i], dst)
    return len(train_idx), len(valid_idx)

counts = {}
for cls in TARGET_CLASSES:
    src_dir = IMAGES_DIR / cls
    n_tr, n_va = train_valid_split(src_dir, TRAIN_DIR, VALID_DIR, valid_frac=0.2, seed=42)
    counts[cls] = (n_tr, n_va)

print("Split summary:", counts)
print("Train classes:", [p.name for p in TRAIN_DIR.iterdir() if p.is_dir()])
print("Valid classes:", [p.name for p in VALID_DIR.iterdir() if p.is_dir()])
```

```python
# --- Cell 3: Datasets & DataLoaders ---
BATCH_SIZE = 16
NUM_WORKERS = 2  # if you see issues, set to 0

# ImageNet normalization for pretrained ResNet
normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225])

train_tfms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    normalize,
])

valid_tfms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    normalize,
])

train_ds = datasets.ImageFolder(str(TRAIN_DIR), transform=train_tfms)
valid_ds = datasets.ImageFolder(str(VALID_DIR), transform=valid_tfms)

print("Classes:", train_ds.classes)
print("Train samples:", len(train_ds), "| Valid samples:", len(valid_ds))

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  num_workers=NUM_WORKERS)
valid_loader = DataLoader(valid_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS)

class_names = train_ds.classes
```

```python
# --- Cell 4: Build the model (transfer learning) ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
weights = ResNet18_Weights.IMAGENET1K_V1
model = resnet18(weights=weights)

# Freeze backbone
for p in model.parameters():
    p.requires_grad = False

# Replace final layer for 2 classes
in_features = model.fc.in_features
model.fc = nn.Linear(in_features, 2)

model = model.to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.fc.parameters(), lr=1e-3)

model
```

```python
# --- Cell 5: Training & evaluation helpers ---
def train_one_epoch(model, loader, optimizer, criterion, device):
    model.train()
    running_loss, running_correct, total = 0.0, 0, 0
    for images, labels in tqdm(loader, desc="Train", leave=False):
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)
        preds = outputs.argmax(1)
        running_correct += (preds == labels).sum().item()
        total += images.size(0)

    return running_loss/total, running_correct/total

@torch.no_grad()
def evaluate(model, loader, criterion, device):
    model.eval()
    running_loss, running_correct, total = 0.0, 0, 0
    all_preds, all_labels = [], []
    for images, labels in tqdm(loader, desc="Valid", leave=False):
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        loss = criterion(outputs, labels)

        running_loss += loss.item() * images.size(0)
        preds = outputs.argmax(1)
        running_correct += (preds == labels).sum().item()
        total += images.size(0)

        all_preds.extend(preds.cpu().numpy().tolist())
        all_labels.extend(labels.cpu().numpy().tolist())

    avg_loss = running_loss/total if total else 0.0
    acc = running_correct/total if total else 0.0
    return avg_loss, acc, np.array(all_labels), np.array(all_preds)
```

```python
# --- Cell 6: Train for a few epochs (CPU-friendly) ---
EPOCHS = 5
best_acc = 0.0

for epoch in range(EPOCHS):
    tl, ta = train_one_epoch(model, train_loader, optimizer, criterion, device)
    vl, va, y_true, y_pred = evaluate(model, valid_loader, criterion, device)
    print(f"Epoch {epoch+1}/{EPOCHS} | train loss {tl:.4f} acc {ta:.3f} | valid loss {vl:.4f} acc {va:.3f}")
    if va > best_acc:
        best_acc = va
        Path("models").mkdir(parents=True, exist_ok=True)
        torch.save(model.state_dict(), "models/wrench_vs_screwdriver_resnet18.pth")
        print("Saved best model -> models/wrench_vs_screwdriver_resnet18.pth")
```

```python
# --- Cell 7: Evaluation summary (confusion matrix & report) ---
labels_range = list(range(len(class_names)))  # [0, 1] for two classes
cm = confusion_matrix(y_true, y_pred, labels=labels_range)
acc = accuracy_score(y_true, y_pred)

print("Accuracy:", f"{acc:.3f}")
print("\nClassification report:\n")
print(classification_report(y_true, y_pred, target_names=class_names))

fig, ax = plt.subplots(figsize=(5, 4))
im = ax.imshow(cm, interpolation='nearest')
ax.set_title("Confusion Matrix")
ax.set_xticks(labels_range); ax.set_xticklabels(class_names, rotation=45, ha="right")
ax.set_yticks(labels_range); ax.set_yticklabels(class_names)
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(j, i, cm[i, j], ha="center", va="center")
ax.set_xlabel("Predicted"); ax.set_ylabel("True")
plt.tight_layout()
plt.show()
```

```python
# --- Cell 8: Single-image inference helper ---
@torch.no_grad()
def predict_image(model, img_path: Path, transform, class_names: List[str]):
    model.eval()
    img = Image.open(img_path).convert("RGB")
    x = transform(img).unsqueeze(0).to(device)
    logits = model(x)
    probs = torch.softmax(logits, dim=1).cpu().numpy().ravel()
    pred_idx = int(np.argmax(probs))
    return class_names[pred_idx], float(probs[pred_idx]), img

# Example usage:
# Update to any real image path below
test_img = Path("data/valid/wrench")
if test_img.exists():
    candidates = [p for p in test_img.glob("*") if p.suffix.lower() in {".jpg",".jpeg",".png"}]
    if candidates:
        label, p, img = predict_image(model, candidates[0], valid_tfms, class_names)
        print(f"Predicted: {label} (p={p:.2f})")
        plt.figure(); plt.imshow(img); plt.axis("off"); plt.show()
    else:
        print("No images found to test in:", test_img)
else:
    print("Set 'test_img' to a real image file path or folder.")
```

```python
# --- Cell 9 (Optional): Scikit-learn baseline with color histograms ---
from sklearn.linear_model import LogisticRegression

def img_paths_and_labels(root: Path, classes: List[str]) -> Tuple[List[Path], List[int]]:
    X, y = [], []
    for idx, cls in enumerate(classes):
        for p in (root/cls).glob("*"):
            if p.suffix.lower() in VALID_EXTS:
                X.append(p)
                y.append(idx)
    return X, y

def rgb_histogram(fp: Path, bins=16):
    img = Image.open(fp).convert("RGB")
    arr = np.array(img)
    feats = []
    for c in range(3):
        hist, _ = np.histogram(arr[:,:,c], bins=bins, range=(0,255), density=True)
        feats.append(hist)
    return np.concatenate(feats)

X_paths, y = img_paths_and_labels(IMAGES_DIR, TARGET_CLASSES)
if len(X_paths) < 10:
    print("Not enough images in data/images for baseline; add more and re-run.")
else:
    X_feats = np.vstack([rgb_histogram(p) for p in tqdm(X_paths, desc="Hist feats")])
    X_tr, X_va, y_tr, y_va = train_test_split(X_feats, y, test_size=0.2, random_state=42, stratify=y)
    clf = LogisticRegression(max_iter=200)
    clf.fit(X_tr, y_tr)
    y_hat = clf.predict(X_va)
    print("Baseline (color histogram) accuracy:", f"{accuracy_score(y_va, y_hat):.3f}")
```
